```python
# Consolidated code to generate the new columns based on heuristic assumptions

def generate_columns(dataset):
    # 1. History_NObeyesdad
    dataset["History_NObeyesdad"] = dataset["family_history_with_overweight"] + "_" + dataset["NObeyesdad"]

    # 2. Obesity_Risk
    conditions_obesity_risk = [
        dataset['NObeyesdad'].str.contains('Obesity'),
        dataset['NObeyesdad'].str.contains('Overweight'),
        dataset['NObeyesdad'].str.contains('Normal_Weight')
    ]
    choices_obesity_risk = ['High', 'Moderate', 'Low']
    dataset['Obesity_Risk'] = np.select(conditions_obesity_risk, choices_obesity_risk, default='Low')

    # 3. Dietary_Habits
    conditions_dietary_habits = [
        (dataset['FAVC'] == 'yes') & (dataset['CH2O'] <= 2),
        (dataset['FAVC'] == 'no') & (dataset['CH2O'] > 2),
    ]
    choices_dietary_habits = ['Poor', 'Good']
    dataset['Dietary_Habits'] = np.select(conditions_dietary_habits, choices_dietary_habits, default='Average')

    # 4. Health_Monitoring
    conditions_health_monitoring = [
        (dataset['SCC'] == 'yes') & (dataset['FAF'] > 1),
        (dataset['SCC'] == 'no') & (dataset['FAF'] <= 1),
    ]
    choices_health_monitoring = ['Attentive', 'Negligent']
    dataset['Health_Monitoring'] = np.select(conditions_health_monitoring, choices_health_monitoring, default='Average')

    # 5. Lifestyle_Score
    conditions_lifestyle_score = [
        (dataset['FAF'] > 1) & (dataset['TUE'] > 1),
        (dataset['FAF'] <= 1) & (dataset['TUE'] <= 1),
    ]
    choices_lifestyle_score = ['Active', 'Sedentary']
    dataset['Lifestyle_Score'] = np.select(conditions_lifestyle_score, choices_lifestyle_score, default='Balanced')

    # 6. Risk_Factors
    conditions_risk_factors = [
        (dataset['family_history_with_overweight'] == 'yes') & (dataset['CALC'].isin(['Frequently', 'Always'])),
        (dataset['family_history_with_overweight'] == 'no') & (dataset['CALC'].isin(['no', 'Sometimes']))
    ]
    choices_risk_factors = ['High_Risk', 'Low_Risk']
    dataset['Risk_Factors'] = np.select(conditions_risk_factors, choices_risk_factors, default='Moderate_Risk')

    # 7. Daily_Routine
    conditions_daily_routine = [
        (dataset['NCP'] > 3) & (dataset['FAF'] > 1),
        (dataset['NCP'] <= 2) & (dataset['FAF'] <= 1)
    ]
    choices_daily_routine = ['Regular', 'Irregular']
    dataset['Daily_Routine'] = np.select(conditions_daily_routine, choices_daily_routine, default='Average')

    # 8. Transport_Physical_Activity
    conditions_transport_activity = [
        dataset['MTRANS'] == 'Walking',
        dataset['MTRANS'] == 'Bike'
    ]
    choices_transport_activity = ['Highly_Active', 'Moderately_Active']
    dataset['Transport_Physical_Activity'] = np.select(conditions_transport_activity, choices_transport_activity, default='Sedentary')

    # 9. Environmental_Factors
    conditions_environmental_factors = [
        dataset['NObeyesdad'].str.contains('Obesity'),
        dataset['NObeyesdad'].str.contains('Overweight')
    ]
    choices_environmental_factors = ['High_Risk', 'Moderate_Risk']
    dataset['Environmental_Factors'] = np.select(conditions_environmental_factors, choices_environmental_factors, default='Low_Risk')
    
    return dataset

# Apply the function to generate columns on the initial dataset
obesity_dataset_updated = generate_columns(obesity_dataset.copy())

# Display the first few rows with the new columns
obesity_dataset_updated.head()
```
